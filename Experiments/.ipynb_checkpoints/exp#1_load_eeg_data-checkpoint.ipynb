```python
# First we need to load EEG data from .edf files

# https://mne.tools/stable/auto_tutorials/intro/10_overview.html#sphx-glr-auto-tutorials-intro-10-overview-py
# https://mne.tools/stable/generated/mne.io.read_raw_edf.html#mne.io.read_raw_edf

folder_path = 'h:\\BCI Project\\eeg-motor-movementimagery-dataset-1.0.0\\files\\S001'
path = os.path.join( folder_path, 'S001R01.edf' )

raw = mne.io.read_raw_edf( path, preload=True )
print(raw.info)
```
